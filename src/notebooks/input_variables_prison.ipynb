```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import pandas as pd
import plotly.express as px
```

```python
data = 'prison'
iter_num = 24
model = 'arima'
colors = px.colors.qualitative.Light24
```

```python
# colors
```

```python
ts_names = pd.read_csv(f'../rolling_window_experiments_transformed/{data}/{model}_fitted_1.csv')
```

```python
ts_names['Description'].values
```

```python
def read_iter(iter_num):
    transformed_input = pd.read_csv(f'../rolling_window_experiments_transformed/{data}/{model}_fitted_transformed_{iter_num}.csv', index_col=0)
    return transformed_input
```

```python
figure = make_subplots(rows=4, cols=4, shared_xaxes = True, subplot_titles = ts_names['Description'].values)

for iter_no in range(1, iter_num + 1):
    iter_data = read_iter(iter_no)
    row = 1 
    col = 1
    for ts in range(0, len(iter_data)):
        if ts == 0:
            show_leg = True
        else:
            show_leg = False
        figure.add_trace(go.Scatter(y = iter_data.iloc[ts, 2:].values, name = f'iteration{iter_no}', showlegend = show_leg, 
                                    legendgroup = f'iteration{iter_no}', marker_color = colors[iter_no-1]), row = row, 
                         col=col)
        col +=1
        if col > 4:
            row +=1
            col = 1
        if row > 4:
            break
```

```python
figure.update_layout(height= 900)
```

```python
figure.write_html('../rolling_window_experiments_transformed/inputs_fitted.html')
```
